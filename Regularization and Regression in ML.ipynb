 Regularization in ML

REgularization is one of the most important concepts of ML. It is a technique to prevent the model from overfitting and underfitting by adding extra information to it.

Sometimes the ML model performs well with the training data but does not perform well with the test data. It means the model is not able to predict the output when deals with unseen data.The problem can be deal with the help of of a regularization technique.

There are 2 terms to define the problem

1)Underfitting

Underfitting is when our model is not able to learn the basic pattern of our dataset. In the case of underfitting model is unable to perform well even on the training data hence we cannot expect it to perform well on the testing data. This is the case when we are suppossed to increase the complexity of the model or add more features to the dataset.

2)Overfitting

Overfitting is phenomenon that occurs when a ML model is able to perform well on training but does not perform well on testing data. This happens when our model memorize the training data instead of the pattern on it.

BIAS

Bias refers to the errors which occurs when we try to fit a statiscal model on real world data which does not fit perfectly well on some mathematical model. High Bias refers to the case when the model is unable to learn the pattern in the data at hand and hence perform poorly. It is mostly under Underfit model

VARIANCE

Variance implies the error value that occurs when we try to make predictions by using data that is not previously seen by model. There is a situation known as high variance that occurs when model is able to perform well on training but does not perform well on testing data.
 How to identify the underfit and overfit?

In underfit MSE high and rscore is low in training===Low features

In overfit MSE 0 and r2score is 0.99 is training but in testing MSE becomes high i.e is overfit situation

# this situation is overfit

```python
# regularuization

from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
l2=Ridge(0.01)
l2.fit(x_train,y_train)
```

```python
l2.score(x_train,y_train)
```

```python
l2.score(x_test,y_test)
```

```python
l2.coef_
```

```python
# for RIGDE we use for loop

for alpha in range(1,5):
    l2=Ridge(alpha)
    l2.fit(x_train,y_train)
    train_score=l2.score(x_train,y_train)
    test_score=l2.score(x_train,y_train)
    print("Alpha:",alpha)
    print("TRAIN_SCORE:", train_score)
    print("TEST_SCORE:", test_score)
    
```

```python
# lasso method
l3=Lasso(1000)
l3.fit(x_train,y_train)
```

```python
l3.score(x_test,y_test)
```

```python
l3.score(x_train,y_train)
```

```python
l3.coef_
```

```python
for i in range(200,500,50):
    l3=Lasso(i)
    l3.fit(x_train,y_train)
    train_score=l3.score(x_train,y_train)
    test_score=l3.score(x_train,y_train)
    print("Alpha:",i)
    print("TRAIN_SCORE:", train_score)
    print("TEST_SCORE:", test_score)
    
    
```

```python
l3.coef_
```

# FINAL RIDGE MODEL
 we are selecting the values from alpha were there is no large different two number like train and test data number 
 
 so we use ALPHA 2 for final model of RIDGE METHOD
 

```python
l2=Ridge(alpha=2)
l2.fit(x_train,y_train)
print(l2.score(x_test,y_test))
```

```python
l2.coef_
```

# FINAL LASSO MODEL

```python
l3=Lasso(alpha=200)
l3.fit(x_train,y_train)
print(l3.score(x_test,y_test))
```
